## Scraping Wiki Page Table Into a Pandas Data Frame.

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
df_pcodes = pd.read_html(url)
```

```python
len(df_pcodes)
```

```python
df_pcodes[0]
```

```python
df_codes = df_pcodes[0]
```

```python
df_codes.head()
```

## Processing the cells that have an assigned borough. Ignoring cells with a borough that is Not assigned.

```python
boroughsNA = df_codes[df_codes['Borough'] == 'Not assigned'].index
```

```python
df_codes.drop(boroughsNA , inplace=True)
```

```python
df_codes.head()
```

# Grouping all Neighborhoods by Postal Code 

```python
df_codes2 = df_codes.groupby('Postal Code').agg(list).reset_index()
```

```python
df_codes2.head(20)
```

## Replacing Not Assigned neighborhoods with their Borough instead

```python
df_codes2['Neighbourhood'] = np.where(df_codes2['Neighbourhood'] == 'Not assigned', df_codes2['Borough'], df_codes2['Neighbourhood'])
```
